```python
import difflib


# Task 1 - Return Brand Name and Model Name where product description is 100% match


# Merge DataFrames based on 'Product Description' with an inner join (same as before)
merged_df = imports_df.merge(data_bank_df, on='Product Description', how='inner')


# Extract 'Brand Name' and 'Model Name' (same as before)
result_task1 = merged_df[['Brand Name', 'Model Name']]


# Task 2 - Return Brand Name and Model Name where product description is up to 90% match


# Get the product descriptions from imports_df that don't have a 100% match in Task 1
not_matched_product_desc = imports_df[~imports_df['Product Description'].isin(merged_df['Product Description'])]


# Initialize an empty DataFrame to store the fuzzy matches
fuzzy_matches_df = pd.DataFrame(columns=['Brand Name', 'Model Name'])


# Iterate over each product description in not_matched_product_desc
for desc in not_matched_product_desc['Product Description']:
    # Skip NaN or missing values in product description
    if pd.notna(desc):
        # Find the best match in the data_bank_df based on partial string matching
        threshold = 0.8 if len(desc) > 15 else 0.9
        best_match = data_bank_df[data_bank_df['Product Description'].str.contains(desc, case=False, na=False, regex=False)]
        if not best_match.empty:
            # If a match is found, append the 'Brand Name' and 'Model Name' to the fuzzy_matches_df
            fuzzy_matches_df = fuzzy_matches_df.append(best_match[['Brand Name', 'Model Name']])


# Concatenate the results of Task 1 and Task 2
result_df = pd.concat([result_task1, fuzzy_matches_df])


# Print the result DataFrame (same as before)
print(result_df)
```

```python
# Task 3 - Find Brand Name from 'Shipper Name' column and find it in Brand table

# Filter rows of the import table where Brand Name and Model Name are not returned from Task 1 and Task 2
rows_without_brand_model = result_df[pd.isnull(result_df['Brand Name']) | pd.isnull(result_df['Model Name'])]

# Initialize a DataFrame to store the results for Task 3
result_task3 = pd.DataFrame(columns=['Product Description', 'Brand Name', 'Model Name'])

# Iterate over each row in the filtered import DataFrame
for _, import_row in rows_without_brand_model.iterrows():
    desc = import_row['Product Description']
    brand_name = None
    model_name = None

    # Try to find Brand Name from 'Shipper Name' column in the 'brand_df' DataFrame
    for _, brand_row in brand_df.iterrows():
        brand_shipper_name = brand_row['Shipper Name']
        if pd.notnull(desc) and pd.notnull(brand_shipper_name):
            # Use SequenceMatcher to calculate similarity ratio
            ratio = difflib.SequenceMatcher(None, str(desc), str(brand_shipper_name)).ratio()
            if ratio >= 0.9:  # Adjust the threshold as needed
                brand_name = brand_row['Brand Name']
                model_name = brand_row['Model Name']
                break  # Stop searching once a match is found

    # Append the result to the 'result_task3' DataFrame
    result_task3 = result_task3.append({
        'Product Description': desc,
        'Brand Name': brand_name,
        'Model Name': model_name
    }, ignore_index=True)

# Concatenate the results from Task 1, Task 2, and Task 3
final_result_df = pd.concat([result_df, result_task3])

# Print the final result DataFrame
print(final_result_df)
```
