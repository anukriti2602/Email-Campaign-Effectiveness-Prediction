##1. Detecting outliers in the "Wage" column involves identifying data points that deviate significantly from the rest of the values. Outliers can be identified using various statistical methods. One common method is the use of the Interquartile Range (IQR) method. Here's the step-by-step process to identify outliers using the IQR method:

Step 1: Calculate the IQR (Interquartile Range)

Sort the "Wage" values in ascending order.
Calculate the first quartile (Q1) which represents the 25th percentile of the data.
Calculate the third quartile (Q3) which represents the 75th percentile of the data.
Calculate the IQR as the difference between Q3 and Q1: IQR = Q3 - Q1.

Step 2: Define the Outlier Boundaries

Calculate the lower bound as Q1 - 1.5 * IQR.
Calculate the upper bound as Q3 + 1.5 * IQR.

Step 3: Identify Outliers

Any data point below the lower bound or above the upper bound is considered an outlier.

```python
# Load the dataset into a DataFrame
data = pd.read_csv("final_cleaned_dataset.csv")  # Replace "final_cleaned_dataset.csv" with the actual file name

# Function to convert wage values to float
def convert_wage_to_float(wage_str):
    wage_str = wage_str.replace('€', '').replace('K', 'e3').replace('M', 'e6')
    return float(wage_str)

# Convert the "Wage" column to numerical format
data["Wage"] = data["Wage"].apply(convert_wage_to_float)

# Get the "Wage" column
wage_values = data["Wage"]

# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = wage_values.quantile(0.25)
Q3 = wage_values.quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = wage_values[(wage_values < lower_bound) | (wage_values > upper_bound)]

# Display the outliers
print("Outliers in the 'Wage' column:")
print(outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify outliers
outliers = wage_values[(wage_values < lower_bound) | (wage_values > upper_bound)]

# Plot the box plot with outliers
plt.figure(figsize=(8, 6))
plt.boxplot(wage_values, vert=False, labels=["Wage"])
plt.scatter(outliers, ["Wage"] * len(outliers), color="r", marker="x", label="Outliers")
plt.xlabel("Wage")
plt.title("Box Plot of Wage with Outliers")
plt.legend()
plt.show()
```

# 3. Difference between normal and student t distribution explain it using 'potential' column.

```python
from scipy.stats import norm, t

# Extract the 'Potential' column
potential_data = data['Potential']

# Calculate mean and standard deviation for potential data
mean = potential_data.mean()
std_dev = potential_data.std()

# Generate data points for the normal distribution based on the mean and standard deviation
normal_dist_data = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
normal_distribution = norm.pdf(normal_dist_data, mean, std_dev)

# Calculate the degrees of freedom for the t-distribution
degrees_of_freedom = len(potential_data) - 1

# Generate data points for the t-distribution based on the degrees of freedom
t_dist_data = np.linspace(t.ppf(0.001, degrees_of_freedom), t.ppf(0.999, degrees_of_freedom), 1000)
t_distribution = t.pdf(t_dist_data, degrees_of_freedom)
```

```python
# Plot the histograms
plt.figure(figsize=(10, 6))
plt.hist(potential_data, bins=30, density=True, alpha=0.7, color='skyblue', label='Potential Data')

# Plot the normal distribution
plt.plot(normal_dist_data, normal_distribution, 'r', label='Normal Distribution')

# Plot the t-distribution
plt.plot(t_dist_data, t_distribution, 'g', label="Student's t-Distribution")

plt.xlabel('Potential')
plt.ylabel('Density')
plt.title('Comparison of Normal Distribution and Student\'s t-Distribution for Potential')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Find the 95%, 90%, and 99%, confidence interval for 'Potential','wage','weight' column.

```python
from scipy import stats

# Confidence level (1 - alpha)
confidence_levels = [0.95, 0.90, 0.99]

# Columns for which to calculate confidence intervals
columns_to_analyze = ['Potential', 'Wage', 'Weight']

# Function to convert 'Weight' values to float (remove 'lbs' string)
def convert_weight_to_float(weight_str):
    try:
        return float(weight_str.replace('lbs', ''))
    except ValueError:
        return np.nan

# Convert 'Weight' column to numeric format
data['Weight'] = data['Weight'].apply(convert_weight_to_float)

# Dictionary to store confidence intervals for each column
confidence_intervals = {}

for column in columns_to_analyze:
    values = data[column].dropna()  # Remove any missing values from the column

    # Calculate the sample mean and standard error for each column
    mean = np.mean(values)
    std_error = stats.sem(values, nan_policy='omit')

    # Calculate the degrees of freedom for t-distribution (only for 'Potential' column)
    df = len(values) - 1 if column == 'Potential' else None

    # Calculate the confidence intervals using the t-distribution (for 'Potential' column) or normal distribution
    intervals = []
    for confidence_level in confidence_levels:
        if column == 'Potential':
            interval = stats.t.interval(confidence_level, df, loc=mean, scale=std_error)
        else:
            interval = stats.norm.interval(confidence_level, loc=mean, scale=std_error)
        intervals.append(interval)

    confidence_intervals[column] = intervals

# Display the confidence intervals
for column, intervals in confidence_intervals.items():
    print(f"Confidence Intervals for '{column}' column:")
    for i, confidence_level in enumerate(confidence_levels):
        print(f"{confidence_level * 100:.0f}% Confidence Interval: {intervals[i]}")
    print()
```

## 6.Proove Central Limit Theorom by using 'potential' column of the game_data.
